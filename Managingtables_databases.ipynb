**Constraints**

used to define rules for the data in a table, prevent insert of invalid data
can be column or table level

Column constraints

    Not NULL    ensure column cannot have a null value
    unique      ensures all values in a column are different
    default     sets a default value for a column if no value specified
    primarykey  A combination of a not null and unique. uniquely identifies  
                each row
    references   ensures referential integrity
    check       ensures that values in a column satisfies a specific condition


Primary Key:

one or multiple columns that uniquely identify each row in a table

Foreign key:
A column (or multiple) that refers to the primary in another table

    1. foreign kkey dies niot need to be unique
